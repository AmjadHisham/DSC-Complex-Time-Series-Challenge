### Example FFT use

* N~number of sample points

* T~Sample spacing ;  f > 1/2*(1/T)

* x~N number of data points from 0 to N*T



When a continuous function, x(t), is sampled at a constant rate, fs samples/second. When bandlimited to ½ fs cycles/second (hertz), its Fourier transform, X(f), is 0 for all |f| ≥ ½ fs. 

```python
from scipy.fftpack import fft
import matplotlib.pyplot as plt
import numpy as np
# Number of samplepoints
N = 600
# sample spacing
T = 1.0 / 800.0
x = np.linspace(0.0, N*T, N)
```

* y~2 sinusodial waves added together with frequencies 50 & 80

* yf~fast fourier transform of y

* xf~N/2 data points from 0 to 1/(2*T)

"If the essential frequency range is limited to B cycles per second, 2B was given by Nyquist as the maximum number of code elements per second that could be unambiguously resolved, assuming the peak interference is less half a quantum step. This rate is generally referred to as signaling at the Nyquist rate and 1/(2B) has been termed a Nyquist interval." (bold added for emphasis; italics from the original)



```python

y = np.sin(50.0 * 2.0*np.pi*x) + 0.5*np.sin(80.0 * 2.0*np.pi*x)
yf = fft(y)
xf = np.linspace(0.0, 1.0/(2*T), N/2)

plt.plot(xf, 2.0/N * np.abs(yf[0:N/2]))
plt.grid()
plt.show()
```
